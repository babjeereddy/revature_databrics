```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Sales': [120, 150, 170, 160, 200],
    'Profit': [30, 40, 45, 38, 55],
    'Category': ['A','B','C','D','E'],
    'Units': [500, 700, 650, 400, 800]
})
df
```

```python
# LINE CHART
#     Shows trend over time
#     Suitable for time-series, performance tracking
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Year'].astype(str), df['Sales'], marker='o')
plt.title("Sales Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Sales': [120, 150, 170, 160, 200],
    'Profit': [30, 40, 45, 38, 55],
    'Expenses': [80, 90, 110, 100, 120]
})

```

```python
df.set_index('Year')[['Sales','Profit','Expenses']].plot(marker='o')
plt.xticks(df['Year'])
plt.title("Multiple Line Trend")
plt.grid(True)
plt.show()
```

```python
# AREA CHART
#     Shows cumulative change
#     Good for demonstrating growth or declin
```

```python
plt.figure(figsize=(10, 2))
plt.fill_between(df['Year'], df['Profit'], color='lightgreen', alpha=0.3)
plt.plot(df['Year'], df['Profit'])
plt.xticks(df['Year'])
plt.title("Profit Area Chart")
plt.xlabel("Year")
plt.ylabel("Profit")
plt.show()
```

```python
df_sorted = df.sort_values('Units', ascending=False)
plt.barh(df_sorted['Category'], df_sorted['Units'])
plt.title("Bar Chart (Sorted Descending)")
plt.xlabel("Category")
plt.ylabel("Units")
```

```python
# PIE CHART
#     Shows percentage share
#     Good for explaining market share, category distribution
plt.pie(df['Units'], labels=df['Category'], autopct='%1.1f%%')

plt.title("Category-wise Units Share")
plt.show()
```

```python
# SCATTER PLOT
#     Shows relationship between two variables
#     Useful for correlation (e.g., sales vs profit)
```

```python
plt.scatter(df['Sales'], df['Profit'])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

```python
# BOX PLOT
#     Shows distribution, median, quartiles, outliers
#     Useful for comparing spread of values
```

```python
plt.boxplot([df['Sales'], df['Profit'], df['Units']], labels=['Sales','Profit','Units'])
plt.title("Box Plot of Metrics")
plt.ylabel("Values")
plt.show()
```

```python
# HISTOGRAM
#     Shows frequency distribution
#     Useful to understand data spread (normal, skewed, etc.)
```

```python
plt.hist(df['Sales'], bins=5)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].plot(df['Year'], df['Sales'], marker='o')
axs[0, 0].set_title("Sales Trend")
axs[0, 0].set_xticks(df['Year'])

df_sorted = df.sort_values('Units', ascending=False)
axs[0, 1].bar(df_sorted['Category'], df_sorted['Units'])
axs[0, 1].set_title("Units by Category (Descending)")

axs[1, 0].pie(df['Units'], labels=df['Category'], autopct="%1.1f%%")
axs[1, 0].set_title("Category Share")

axs[1, 1].scatter(df['Sales'], df['Profit'])
axs[1, 1].set_title("Sales vs Profit")
axs[1, 1].set_xlabel("Sales")
axs[1, 1].set_ylabel("Profit")
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
