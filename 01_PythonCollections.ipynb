```python
# A defaultdict is a subclass of Pythonâ€™s builtâ€‘in dict, in the collections module.
# It automatically provides a default value for missing keys â€” 
# so you can access or append to new keys without getting a KeyError.
```

```python
from collections import defaultdict
inventory = defaultdict(int)
inventory['apple']=10
print(inventory['apple'])
print(inventory['banana'])
```

```python
from collections import defaultdict
inventory = defaultdict(lambda : 'out of stock')
inventory['apple']=10
print(inventory['apple'])
print(inventory['banana'])
```

```python
from collections import defaultdict

# Create a defaultdict where missing keys start with 0
counts = defaultdict(int)

words = ["apple", "banana", "apple", "orange", "banana", "apple"]

# Count how many times each fruit appears
for fruit in words:
    counts[fruit] += 1

print(counts)
```

```python
# A ChainMap is a data structure from the collections module that 
# groups multiple dictionaries (or mappings) into a single, viewâ€‘like object.

# When you search for a key, ChainMap looks through its dictionaries one by one, in order, 
# until it finds the key.
# Itâ€™s like stacking several dictionaries on top of each other to form a single 
# logical mapping â€” without merging them physically.
```

```python
from collections import ChainMap


defaults = {'theme': 'Light', 'language': 'English'}
user_settings = {'theme': 'Dark', 'font': 'Arial'}

# Chain them together
settings = ChainMap(user_settings, defaults)
print(settings)
```

```python
print(settings['theme'])     
print(settings['language'])   
print(settings['font'])       
```

```python
settings['theme'] = 'Blue'
print(user_settings)  
#The original underlying dictionaries are updated â€” ChainMap doesnâ€™t copy data.

```

```python
#Temporary Overrides


from collections import ChainMap

defaults = {"user": "Guest", "theme": "Light"}
temp = defaults.copy()
temporary_view = ChainMap(temp, defaults)

temporary_view["theme"] = "Dark"
print(temporary_view["theme"])  # Dark
print(defaults["theme"])        # Light (unchanged)
```

```python
# A Counter is a specialized dictionary subclass from Pythonâ€™s collections module,
# designed for counting hashable objects.
# It automatically keeps track of how many times each element appears â€” perfect for word frequencies, votes
```

```python
from collections import Counter
fruits = ["apple", "banana", "apple", "orange", "banana", "apple"]
count = Counter(fruits)
print(count)
```

```python
print(count["apple"])   
print(count["grape"])   
```

```python
count["banana"] += 1
count["grape"] = 5
print(count)
```

```python
word_count = Counter("mississippi")
print(word_count)
```

```python
sentence = "life is short and python is fun"
words = sentence.split()
word_counter = Counter(words)
print(word_counter.most_common(3))
```

```python
#combining counters
c1 = Counter(a=3, b=1)
c2 = Counter(a=1, b=2, c=3)

print(c1 + c2)  
print(c1 - c2)  # Subtract counts (no negatives)
```

```python
# A Context Manager in Python is a protocol that sets up a temporary runtime context,
# executes a block of code within it, and then cleans up automatically when the block exits.

# Itâ€™s most famously used with the with statement.
```

```python
file = open("data.txt", "r")
content = file.read()
file.close()  # must remember to do this manually
print(content)
```

```python
with open("data.txt", "r") as file:
    content = file.read()
# file gets closed automatically 
```

```python
from contextlib import contextmanager
import socket

@contextmanager
def open_socket(host, port):
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    try:
        print("ðŸ”¹ Connecting...")
        s.connect((host, port))
        yield s
    finally:
        print("ðŸ”¸ Closing...")
        s.close()

# Usage
with open_socket("example.com", 80) as s:
    s.sendall(b"HEAD / HTTP/1.0\r\nHost: example.com\r\n\r\n")
    print(s.recv(200))
```
